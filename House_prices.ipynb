```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/MaciejTeleon/TensorFlow/main/kc_house_data.csv'
df = pd.read_csv(url)
```

```python
df.isnull().sum()
```

```python
df.describe().transpose()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'])
```

```python
sns.countplot(df['bedrooms'])
```

```python
df.corr()['price'].sort_values()
```

```python
sns.boxplot(x='bedrooms', y = 'price', data = df)
```

```python
sns.scatterplot(x = 'bathrooms', y = 'price', data = df)
```

```python
plt.figure(figsize=(15,20))
sns.scatterplot(x='price', y ='long', data = df)
```

```python
plt.figure(figsize=(15,20))
sns.scatterplot(x='price', y ='lat', data = df)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='long', y='lat', data=df, hue='price')
```

```python
df.sort_values('price', ascending=False).head(20)
```

```python
df_cut = df.loc[df['price'] <= 3000000]
```

```python
len(df_cut)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='long', y='lat', data=df_cut, hue='price', edgecolor = None, alpha=0.2, palette='RdYlGn')
```
